# Определение стоимости автомобилей

Сервис по продаже автомобилей с пробегом «Не бит, не крашен» разрабатывает приложение для привлечения новых клиентов. В нём можно быстро узнать рыночную стоимость своего автомобиля. В вашем распоряжении исторические данные: технические характеристики, комплектации и цены автомобилей. Вам нужно построить модель для определения стоимости. 

Заказчику важны:

- качество предсказания;
- скорость предсказания;
- время обучения.

__Признаки__
* DateCrawled — дата скачивания анкеты из базы
* VehicleType — тип автомобильного кузова
* RegistrationYear — год регистрации автомобиля
* Gearbox — тип коробки передач
* Power — мощность (л. с.)
* Model — модель автомобиля
* Kilometer — пробег (км)
* RegistrationMonth — месяц регистрации автомобиля
* FuelType — тип топлива
* Brand — марка автомобиля
* NotRepaired — была машина в ремонте или нет
* DateCreated — дата создания анкеты
* NumberOfPictures — количество фотографий автомобиля
* PostalCode — почтовый индекс владельца анкеты (пользователя)
* LastSeen — дата последней активности пользователя <br>
__Целевой признак__ <br>
* Price — цена (евро)

```python
import time as t
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from datetime import datetime
from catboost import CatBoostRegressor
import lightgbm as lgb
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

## Подготовка данных

Моя функция с предыдущих уроков для отслеживания аномалий: 

```python
def check_dataset(label : str, data : pd.DataFrame) -> None:
    """Вывод важных параметров исследуемого датасета."""
    print(f'________ Рассматриваем датасет {label}')
    print(f'Размерность: {data.shape}')
    print(f'Дублирующих записей: {data.duplicated().sum() / len(data)} %')
    print('Пример датасета:')
    display(data.head())
    res = pd.DataFrame()
    res['type'] = data.dtypes
    res['Uniq'] = 1 - data.nunique() / len(data)
    res['NaN'] = data.isna().sum() / len(data)
    res = res.merge(data.describe(percentiles = [.01, .25, .5, .75, .99]).T, 
                    how = 'left', 
                    left_index = True, 
                    right_index = True)
    res = res.sort_values(by = 'type')
    columns_cat = res.loc[res['type'] == 'object'].index
    res = res.style.background_gradient(axis=1, subset = ['Uniq', 'NaN'], cmap='Blues', vmin = 0.01, vmax = 1)
    display(res)
    for k, i in enumerate(columns_cat):
        if k == 0:
            print('Уникальные значения категориальных данных:')
            print()
        print('Признак ', i, ':', data[i].unique())
    
```

Загрузим датасет:

```python
dateparse = lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S')
try:
    data = pd.read_csv('autos.csv', 
                       parse_dates=['DateCrawled', 'DateCreated', 'LastSeen'], 
                       date_parser=dateparse)
except:
    data = pd.read_csv('https://code.s3.yandex.net/datasets/autos.csv', 
                       parse_dates=['DateCrawled', 'DateCreated', 'LastSeen'], 
                       date_parser=dateparse)
```

Проверка: 

```python
check_dataset('', data)
```

Есть неинформационный столбцы, удалим их:

```python
data = data.drop(columns = ['DateCrawled','DateCreated','PostalCode','LastSeen','NumberOfPictures'])
```

Отфильтруем аномалии и пропуски: 

```python
data = data[(data['Model'].notna()) &
             (data['Gearbox'].notna()) &
             (data['VehicleType'].notna()) &
             (data['FuelType'].notna()) &
             ((data['Power'] > 10) & (data['Power'] < 500) | (data['Power'] == 0)) &
             (data['Price'] > 0) &
             (data['RegistrationYear'] > 1950)]
```

Пометим значение, когда пользователь не указывает был ли ремонт, что ремонта не было.

```python
data['NotRepaired'] = data['NotRepaired'].fillna('no')
```

Приведем столбец с ремонтом к целочисленному виду:

```python
data['NotRepaired'] = data['NotRepaired'].map({'yes':1, 'no':0})
```

Есть явные дубликаты, удалим их. 

```python
data = data.drop_duplicates()
```

```python
def get_type_segment_age(x : int):
    if x < 1990:
        return 'retro'
    if x < 2000:
        return 'old'
    if x < 2010:
        return 'transition'
    if x < 2017:
        return 'now'
    return 'new'
```

```python
data['type_segment_age'] = data['RegistrationYear'].apply(get_type_segment_age)
```

Проверим датасет ещё раз:

```python
check_dataset('', data)
```

Все аномалии и пропуски ушли. 

Отработаем категориальный признак с помощью LabelEncoder:

```python
label_encode = LabelEncoder()

for column in data.select_dtypes(include='object').columns:
    label_encode.fit(data[column].astype('str'))
    data[column] = label_encode.transform(data[column].astype('str'))
```

Проверим корреляцию в данных:

```python
plt.figure(figsize = (10, 10))
sns.heatmap(data.corr(), xticklabels=data.corr().columns, yticklabels=data.corr().columns, cmap='RdYlGn', center=0,
                    annot=True)
```

Есть средние положительные взаимосвязи, но объединять признаки здесь, не вижу смысла, оставим как есть. 

Прекрасно, провели предобработку данных, привели данные к анализу. 

## Обучение моделей

Первым делом разделим данные на тренировочную и валидационную выборку:

```python
features, features_test, target, target_test = train_test_split(data.drop(columns = ['Price']),
                                                                data['Price'],
                                                                test_size = 0.1,
                                                                random_state = 12345)
features_train, features_valid, target_train, target_valid = train_test_split(features,
                                                                              target,
                                                                              test_size = 0.2,
                                                                              random_state = 12345)
```

Проведем масштабирование численных значений:

```python
scaler = StandardScaler()
numeric = ['RegistrationYear','Power','Kilometer']
scaler.fit(features_train[numeric])
features_train[numeric] = scaler.transform(features_train[numeric])
```

И так моя стратегия такая:  <br>
Будем сразу тренировать все модели, изменяя гиперпараметры и записывать в новый датасет, их гиперпараметры, время обучения и метрику. <br>
Для этого написан таймер:

```python
class Timer:
    """Таймер обучения модели."""
    def __init__(self) -> None:
        self.time_start = t.time()

    def get_process_time(self) -> t:
        return t.time() - self.time_start
```

Функция для добавления строки в результирующий датасет: 

```python
def get_row_to_result(model,
                      time : float, 
                      target_valid : pd.Series,
                      features_valid : pd.DataFrame) -> None:
    global result
    features_valid[numeric] = scaler.transform(features_valid[numeric])
    time_predict = Timer()
    predict = model.predict(features_valid)
    result = result.append({"name_model" : model.__class__.__name__ , 
                            "parametrs" : model.get_params(), 
                            "time_train" : time,
                            "time_predict" : time_predict.get_process_time(),
                            "score" : mean_squared_error(target_valid, predict, squared = False)},
                            ignore_index = True)
    
```

Создадим результурующий датасет: 

```python
result = pd.DataFrame(columns = ['name_model', 'parametrs', 'time_train', 'score'])
```

Линейная регрессия:

```python
model = LinearRegression()
timer = Timer()
model.fit(features_train, target_train)
time = timer.get_process_time()
get_row_to_result(model, time, target_valid, features_valid)
```

Случайный лес:

```python
for depth in range(1, 52, 5):
    model = DecisionTreeRegressor(max_depth = depth)
    timer = Timer()
    model.fit(features_train, target_train)
    time = timer.get_process_time()
    get_row_to_result(model, time, target_valid, features_valid)
```

Лес решений:

```python
for est in range(100, 201, 50):
    for depth in range(10, 31, 10):
        model = RandomForestRegressor(max_depth = depth, n_estimators = est)
        timer = Timer()
        model.fit(features_train, target_train)
        time = timer.get_process_time()
        get_row_to_result(model, time, target_valid, features_valid)
```

CatBoost:

```python
categorical = ['Brand','Model','VehicleType','Gearbox','FuelType', 'type_segment_age']
```

```python
for iterat in range(100, 201, 100):
    for depth in range(5, 17, 5):
        model = CatBoostRegressor(loss_function='RMSE', 
                                iterations=iterat, 
                                learning_rate=0.2, 
                                thread_count=-1, 
                                random_state=18, 
                                max_depth=depth, 
                                l2_leaf_reg=3.5,
                                cat_features=categorical)
        timer = Timer()
        model.fit(features_train, target_train)
        time = timer.get_process_time()
        get_row_to_result(model, time, target_valid, features_valid)
```

LightGBM:

```python
for est in range(50, 101, 50):
    for depth in range(10, 21, 10):
        model = lgb.LGBMRegressor(task='train', 
                                boosting_type='gbdt', 
                                objective='regression', 
                                metric='rmse', 
                                learning_rate=0.5, 
                                verbose=200,
                                max_depth = depth,
                                n_estimators = est);
        timer = Timer()
        model.fit(features_train, target_train)
        time = timer.get_process_time()
        get_row_to_result(model, time, target_valid, features_valid)
```

## Анализ моделей

И так вот наша результирующая таблица: 

```python
result
```

Сделаем красивую таблицу с результами:

```python
result_tmp = pd.DataFrame()

for name_model in result['name_model'].unique():
    index = result.loc[result['name_model'] == name_model ,'score'].idxmin(axis=0)
    result_tmp = result_tmp.append(result.iloc[index, :])

result_tmp = result_tmp.merge(result.groupby('name_model').agg({'time_train' : 'sum'}).rename(columns = {'time_train' : 'main_time'}), 
                              how = 'left', 
                              on = 'name_model')
```

```python
for parametr in ['max_depth', 'n_estimators', 'iterations']:
    for i in result.index:
        try:
            result_tmp.loc[i, parametr] = result_tmp.parametrs[i][parametr]
        except:
            pass
result2 = result_tmp.drop(columns = ['parametrs'])
result2.sort_values(by = 'score')
```

И так, поясню, данная таблица показывает не только лучшую метрику модели с параметрами и время на обучения, а также совокупное время поиска лучшей модели.

Как мы видим, лучшую метрику дал LinearRegression c RMSE = 4082 и время обучения превосходны.

Но проведя контрольную, оказалась, хуже, контрольную проводим на CatBoost.


Котрольный замер с моделью:

```python
model = CatBoostRegressor(loss_function='RMSE', 
                            iterations=200, 
                            learning_rate=0.2, 
                            thread_count=-1, 
                            random_state=18, 
                            max_depth=15, 
                            l2_leaf_reg=3.5,
                            cat_features=categorical)
timer = Timer()
model.fit(features_train, target_train)
time = timer.get_process_time()
timer_predict = Timer()
features_test[numeric] = scaler.transform(features_test[numeric])
predict = model.predict(features_test)
time_predict = timer_predict.get_process_time()
score = mean_squared_error(predict, target_test, squared = False)
print(f'RMSE = {score}, время обучения = {time}, время предсказания = {time_predict}')
```
